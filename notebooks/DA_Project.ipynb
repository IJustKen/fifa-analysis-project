# KNN Imputation and Median imputation

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer, SimpleImputer

# 1. DROP columns with >50% missing
cols_to_drop = ['club_loaned_from', 'nation_team_id', 'nation_position',
                'nation_jersey_number', 'player_tags', 'player_traits', 'club_joined_date']
df = df.drop(columns=cols_to_drop, errors='ignore')

print(f"Dropped {len(cols_to_drop)} columns")

# 2. Separate goalkeepers and outfield players
is_gk = df['player_positions'].str.contains('GK', na=False)
df_gk = df[is_gk].copy()
df_outfield = df[~is_gk].copy()

print(f"Outfield: {len(df_outfield)}, GK: {len(df_gk)}")

# 3. KNN Imputation for CORE SKILLS (11% missing)
skill_cols = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']

# For outfield players
print("\n🔧 KNN Imputation for outfield player skills...")
# Select features for KNN context (skills + related attributes)
outfield_features_for_knn = skill_cols + ['overall', 'potential', 'age', 'height_cm', 'weight_kg']
outfield_knn_data = df_outfield[outfield_features_for_knn].copy()

# KNN Imputer (n_neighbors=5 is default, adjust if needed)
knn_imputer = KNNImputer(n_neighbors=5, weights='distance')
outfield_imputed = knn_imputer.fit_transform(outfield_knn_data)

# Put back into dataframe
df_outfield[outfield_features_for_knn] = outfield_imputed

print(f"✓ Outfield skills imputed with KNN")

# For goalkeepers (simpler - fewer GKs so median is fine)
print("\n🔧 Median imputation for GK skills...")
for col in skill_cols:
    df_gk[col] = df_gk[col].fillna(df_gk[col].median())

print(f"✓ GK skills imputed with median")

# 4. MEDIAN Imputation for low-missing columns (<1%)
print("\n🔧 Median imputation for financial/club data...")

low_missing_cols = ['value_eur', 'wage_eur', 'release_clause_eur',
                    'league_id', 'league_level', 'club_team_id',
                    'club_jersey_number', 'club_contract_valid_until_year',
                    'club_position', 'nation_jersey_number']

# Numeric columns
numeric_low_missing = ['value_eur', 'wage_eur', 'release_clause_eur',
                       'league_id', 'league_level', 'club_team_id',
                       'club_jersey_number', 'club_contract_valid_until_year']

for col in numeric_low_missing:
    if col in df_outfield.columns:
        df_outfield[col] = df_outfield[col].fillna(df_outfield[col].median())
    if col in df_gk.columns:
        df_gk[col] = df_gk[col].fillna(df_gk[col].median())

# Categorical columns (club_name, league_name, etc.)
categorical_cols = ['league_name', 'club_name', 'club_position']
for col in categorical_cols:
    if col in df_outfield.columns:
        df_outfield[col] = df_outfield[col].fillna('Free_Agent')
    if col in df_gk.columns:
        df_gk[col] = df_gk[col].fillna('Free_Agent')

print(f"✓ Financial/club data imputed with median")

# 5. Combine back
df_clean = pd.concat([df_outfield, df_gk], ignore_index=True)

# 6. Verify
print(f"\n{'='*60}")
print(f"FINAL CLEAN DATASET")
print(f"{'='*60}")
print(f"Total rows: {len(df_clean)}")
print(f"Total columns: {df_clean.shape[1]}")
print(f"\nRemaining missing values:")
remaining_nulls = df_clean.isnull().sum()
remaining_nulls = remaining_nulls[remaining_nulls > 0].sort_values(ascending=False)
if len(remaining_nulls) > 0:
    print(remaining_nulls)
else:
    print("✅ No missing values remaining!")

# Save cleaned dataset
df_clean.to_csv('/content/drive/MyDrive/FIFA_Project/data/fifa22_cleaned.csv', index=False)
print(f"\n✅ Cleaned dataset saved!")
```

```python
from scipy import stats

# Define hypotheses
H0 = "There is no difference in average wages between Top 5 leagues and other leagues"
H1 = "Top 5 leagues have different (higher) average wages than other leagues"

print(f"Null Hypothesis (H0): {H0}")
print(f"Alternative Hypothesis (H1): {H1}\n")

# Top 5 leagues vs others
top_5_leagues = ['Premier League', 'La Liga', 'Serie A', 'Bundesliga', 'Ligue 1']
df['is_top5_league'] = df['league_name'].isin(top_5_leagues)

top5_wages = df[df['is_top5_league']]['wage_eur'].dropna()
other_wages = df[~df['is_top5_league']]['wage_eur'].dropna()

# Independent t-test
t_stat, p_value = stats.ttest_ind(top5_wages, other_wages)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.6f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print(f"✓ Decision: REJECT H0 (p = {p_value:.6f} < {alpha})")
    print(f"Conclusion: {H1}")
else:
    print(f"✗ Decision: FAIL TO REJECT H0 (p = {p_value:.6f} >= {alpha})")
    print(f"Conclusion: {H0}")
```

```python
from scipy.stats import f_oneway

# Define hypotheses
H0 = "The average wage is the same across all positions (Attack, Midfield, Defense, GK)"
H1 = "At least one position has a different average wage"

print(f"Null Hypothesis (H0): {H0}")
print(f"Alternative Hypothesis (H1): {H1}\n")

# Group positions
positions = ['Attack', 'Midfield', 'Defense', 'GK']
position_groups = []

for pos in positions:
    if pos == 'Attack':
        group = df[df['player_positions'].str.contains('ST|LW|RW|CF', na=False)]['wage_eur'].dropna()
    elif pos == 'Midfield':
        group = df[df['player_positions'].str.contains('CM|CAM|CDM', na=False)]['wage_eur'].dropna()
    elif pos == 'Defense':
        group = df[df['player_positions'].str.contains('CB|LB|RB|LWB|RWB', na=False)]['wage_eur'].dropna()
    else:
        group = df[df['player_positions'].str.contains('GK', na=False)]['wage_eur'].dropna()
    position_groups.append(group)

# One-way ANOVA
f_stat, p_value = f_oneway(*position_groups)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.6f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print(f"✓ Decision: REJECT H0 (p = {p_value:.6f} < {alpha})")
    print(f"Conclusion: {H1}")
else:
    print(f"✗ Decision: FAIL TO REJECT H0 (p = {p_value:.6f} >= {alpha})")
    print(f"Conclusion: {H0}")
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Define hypotheses
H0 = "Player position and preferred foot are independent (no association)"
H1 = "Player position and preferred foot are dependent (there is an association)"

print(f"Null Hypothesis (H0): {H0}")
print(f"Alternative Hypothesis (H1): {H1}\n")

# Create simplified position
df['position_simple'] = df['player_positions'].apply(
    lambda x: 'Left Wing' if any(p in str(x) for p in ['LW'])
    else 'Right Wing' if any(p in str(x) for p in ['RW'])
    else 'Center Forward' if any(p in str(x) for p in ['CF','ST'])
    else 'Midfield' if any(p in str(x) for p in ['CM', 'CAM', 'CDM','RM','LM'])
    else 'Defense' if any(p in str(x) for p in ['CB', 'LB', 'RB'])
    else 'GK'
)

# Contingency table
contingency_table = pd.crosstab(df['position_simple'], df['preferred_foot'])
print("Contingency Table:")
print(contingency_table)
print()

# Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.6f}")
print(f"Degrees of freedom: {dof}")

# Decision
alpha = 0.05
if p_value < alpha:
    print(f"✓ Decision: REJECT H0 (p = {p_value:.6f} < {alpha})")
    print(f"Conclusion: {H1}")
else:
    print(f"✗ Decision: FAIL TO REJECT H0 (p = {p_value:.6f} >= {alpha})")
    print(f"Conclusion: {H0}")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Define hypotheses
H0 = "The proportion of fast attackers (pace > 75) is equal to 60%"
H1 = "The proportion of fast attackers (pace > 75) is not equal to 60%"

print(f"Null Hypothesis (H0): {H0}")
print(f"Alternative Hypothesis (H1): {H1}\n")

# Get attackers
attackers = df[df['player_positions'].str.contains('ST|LW|RW|CF', na=False)]
fast_attackers = attackers[attackers['pace'] > 75]

count = len(fast_attackers)
nobs = len(attackers)
value = 0.60  # Null hypothesis value

# Z-test
z_stat, p_value = proportions_ztest(count, nobs, value)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.6f}")
print(f"Actual proportion: {count/nobs:.2%}")

# Decision
alpha = 0.05
if p_value < alpha:
    if z_stat > 0:
        print(f"✓ Decision: REJECT H0 (p = {p_value:.6f} < {alpha})")
        print(f"Conclusion: Proportion is significantly HIGHER than 60%")
    else:
        print(f"✓ Decision: REJECT H0 (p = {p_value:.6f} < {alpha})")
        print(f"Conclusion: Proportion is significantly LOWER than 60% (actual: {count/nobs:.2%})")
else:
    print(f"✗ Decision: FAIL TO REJECT H0 (p = {p_value:.6f} >= {alpha})")
    print(f"Conclusion: {H0}")
```

```python
import json
from scipy import stats
from scipy.stats import f_oneway, chi2_contingency
from statsmodels.stats.proportion import proportions_ztest
import pandas as pd

# Initialize list to store all test results
hypothesis_tests = []

# ============================================
# TEST 1: Independent T-Test (Top 5 Leagues vs Others)
# ============================================
H0 = "There is no difference in average wages between Top 5 leagues and other leagues"
H1 = "Top 5 leagues have different (higher) average wages than other leagues"

top_5_leagues = ['Premier League', 'La Liga', 'Serie A', 'Bundesliga', 'Ligue 1']
df['is_top5_league'] = df['league_name'].isin(top_5_leagues)

top5_wages = df[df['is_top5_league']]['wage_eur'].dropna()
other_wages = df[~df['is_top5_league']]['wage_eur'].dropna()

t_stat, p_value = stats.ttest_ind(top5_wages, other_wages)

hypothesis_tests.append({
    'test_id': 1,
    'test_name': 'Independent T-Test',
    'test_type': 't-test',
    'comparison': 'Top 5 Leagues vs Other Leagues',
    'variable': 'Wage (EUR)',
    'null_hypothesis': H0,
    'alternative_hypothesis': H1,
    'statistic_name': 'T-statistic',
    'statistic_value': round(float(t_stat), 4),
    'p_value': round(float(p_value), 6),
    'alpha': 0.05,
    'significant': bool(p_value < 0.05),
    'decision': 'Reject H0' if p_value < 0.05 else 'Fail to reject H0',
    'interpretation': 'Top 5 leagues pay significantly higher wages' if p_value < 0.05 else 'No significant difference in wages',
    'group1_mean': round(float(top5_wages.mean()), 2),
    'group2_mean': round(float(other_wages.mean()), 2),
    'group1_name': 'Top 5 Leagues',
    'group2_name': 'Other Leagues',
    'sample_size_group1': int(len(top5_wages)),
    'sample_size_group2': int(len(other_wages))
})

# ============================================
# TEST 2: ANOVA (Wages Across Positions)
# ============================================
H0 = "The average wage is the same across all positions"
H1 = "At least one position has a different average wage"

positions = ['Attack', 'Midfield', 'Defense', 'GK']
position_groups = []
position_means = {}

for pos in positions:
    if pos == 'Attack':
        group = df[df['player_positions'].str.contains('ST|LW|RW|CF', na=False)]['wage_eur'].dropna()
    elif pos == 'Midfield':
        group = df[df['player_positions'].str.contains('CM|CAM|CDM', na=False)]['wage_eur'].dropna()
    elif pos == 'Defense':
        group = df[df['player_positions'].str.contains('CB|LB|RB|LWB|RWB', na=False)]['wage_eur'].dropna()
    else:
        group = df[df['player_positions'].str.contains('GK', na=False)]['wage_eur'].dropna()
    position_groups.append(group)
    position_means[pos] = round(float(group.mean()), 2)

f_stat, p_value = f_oneway(*position_groups)

hypothesis_tests.append({
    'test_id': 2,
    'test_name': 'One-Way ANOVA',
    'test_type': 'f-test',
    'comparison': 'Wages Across Positions',
    'variable': 'Wage (EUR)',
    'null_hypothesis': H0,
    'alternative_hypothesis': H1,
    'statistic_name': 'F-statistic',
    'statistic_value': round(float(f_stat), 4),
    'p_value': round(float(p_value), 6),
    'alpha': 0.05,
    'significant': bool(p_value < 0.05),
    'decision': 'Reject H0' if p_value < 0.05 else 'Fail to reject H0',
    'interpretation': 'At least one position has significantly different wages' if p_value < 0.05 else 'No significant difference in wages across positions',
    'group_means': position_means,
    'groups': positions
})

# ============================================
# TEST 3: Chi-Square (Position vs Preferred Foot)
# ============================================
H0 = "Player position and preferred foot are independent (no association)"
H1 = "Player position and preferred foot are dependent (there is an association)"

df['position_simple'] = df['player_positions'].apply(
    lambda x: 'Left Wing' if any(p in str(x) for p in ['LW'])
    else 'Right Wing' if any(p in str(x) for p in ['RW'])
    else 'Center Forward' if any(p in str(x) for p in ['CF','ST'])
    else 'Midfield' if any(p in str(x) for p in ['CM', 'CAM', 'CDM','RM','LM'])
    else 'Defense' if any(p in str(x) for p in ['CB', 'LB', 'RB'])
    else 'GK'
)

contingency_table = pd.crosstab(df['position_simple'], df['preferred_foot'])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Calculate percentages for interpretation
contingency_pct = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100

hypothesis_tests.append({
    'test_id': 3,
    'test_name': 'Chi-Square Test of Independence',
    'test_type': 'chi-square',
    'comparison': 'Position vs Preferred Foot',
    'variable': 'Categorical Association',
    'null_hypothesis': H0,
    'alternative_hypothesis': H1,
    'statistic_name': 'Chi-square statistic',
    'statistic_value': round(float(chi2), 4),
    'p_value': round(float(p_value), 6),
    'degrees_of_freedom': int(dof),
    'alpha': 0.05,
    'significant': bool(p_value < 0.05),
    'decision': 'Reject H0' if p_value < 0.05 else 'Fail to reject H0',
    'interpretation': 'Significant association - Defenders have higher proportion of left-footed players (32.8%)' if p_value < 0.05 else 'No significant association',
    'contingency_table': contingency_table.to_dict(),
    'contingency_percentages': {k: {k2: round(v2, 2) for k2, v2 in v.items()} for k, v in contingency_pct.to_dict().items()}
})

# ============================================
# TEST 4: Z-Test (Proportion of Fast Attackers)
# ============================================
H0 = "The proportion of fast attackers (pace > 75) is equal to 60%"
H1 = "The proportion of fast attackers is not equal to 60%"

attackers = df[df['player_positions'].str.contains('ST|LW|RW|CF', na=False)]
fast_attackers = attackers[attackers['pace'] > 75]

count = len(fast_attackers)
nobs = len(attackers)
value = 0.60

z_stat, p_value = proportions_ztest(count, nobs, value)
actual_proportion = count / nobs

hypothesis_tests.append({
    'test_id': 4,
    'test_name': 'One-Sample Z-Test for Proportion',
    'test_type': 'z-test',
    'comparison': 'Proportion of Fast Attackers',
    'variable': 'Proportion (pace > 75)',
    'null_hypothesis': H0,
    'alternative_hypothesis': H1,
    'statistic_name': 'Z-statistic',
    'statistic_value': round(float(z_stat), 4),
    'p_value': round(float(p_value), 6),
    'alpha': 0.05,
    'significant': bool(p_value < 0.05),
    'decision': 'Reject H0' if p_value < 0.05 else 'Fail to reject H0',
    'interpretation': f'Actual proportion ({actual_proportion:.2%}) is significantly lower than 60%' if (p_value < 0.05 and z_stat < 0) else f'Actual proportion ({actual_proportion:.2%}) is significantly higher than 60%' if (p_value < 0.05 and z_stat > 0) else 'No significant difference from 60%',
    'hypothesized_proportion': 0.60,
    'actual_proportion': round(float(actual_proportion), 4),
    'sample_size': int(nobs),
    'successes': int(count)
})

# ============================================
# EXPORT TO JSON
# ============================================
output_data = {
    'title': 'FIFA 23 Hypothesis Testing Results',
    'description': 'Statistical hypothesis tests performed on player data',
    'total_tests': len(hypothesis_tests),
    'significance_level': 0.05,
    'tests': hypothesis_tests
}

# Save to JSON file
with open('hypothesis_tests_results.json', 'w') as f:
    json.dump(output_data, f, indent=2)

print("✅ Hypothesis tests exported to hypothesis_tests_results.json")
print(f"Total tests: {len(hypothesis_tests)}")
print("\nPreview:")
print(json.dumps(output_data, indent=2)[:1000] + "...")
```
